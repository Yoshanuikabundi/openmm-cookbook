Unit cell is from crystallography. We want sphere-like PBCs, so we need to change it:

```python
import numpy as np
from scipy.spatial.distance import cdist
from openmm import unit

points = np.asarray(lysozyme_pos.value_in_unit(unit.nanometer))
dists = unit.Quantity(cdist(points, points, 'euclidean'), unit.nanometer)
max_len = dists.max()
max_len
```
